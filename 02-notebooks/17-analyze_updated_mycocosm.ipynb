```python
if RUN_PREPROCESS:
    os.system(f'cat {UPDATED_DIR}/outputs/*csv > combined.csv')
    updated_df = pd.read_csv('combined.csv')
    initial_df = pd.read_csv(f'../../02-OUTPUT/07-tile_mycocosm/03-clustered_90/output_combined.csv')
    merged_df = updated_df.merge(initial_df,on='tile_id')
    (merged_df
     .rename(columns={'predictions_x':'new_predictions',
                      'predictions_y':'old_predictions',
                      'uncertainty_x':'new_uncertainty',
                      'uncertainty_y':'old_uncertainty'})
     .to_pickle('updated_mycocosm_results.pkl'))
else:
    merged_df = pd.read_pickle('../../02-OUTPUT/xx-sandbox/updated_mycocosm_results.pkl')
    merged_df['new_predictions'] = merged_df['new_predictions'].astype(float)
    merged_df['new_uncertainty'] = merged_df['new_uncertainty'].astype(float)
    merged_df = merged_df.round(3)
merged_df
```

```python
fig = plt.figure(figsize=(2.5,2.5))
gs = fig.add_gridspec(6, 6)

test = merged_df
ax_joint = fig.add_subplot(gs[1:6, 0:5])
g = sns.scatterplot(data=test,x='old_predictions',y='old_uncertainty',s=3,color='lightgrey',edgecolor=None,ax=ax_joint,label='Initial')
g = sns.scatterplot(data=test,x='new_predictions',y='new_uncertainty',s=3,color='cadetblue',edgecolor=None,ax=ax_joint,label='Updated')
g.xaxis.set_major_locator(ticker.MultipleLocator(base=1.0))
g.set(xlabel='Predicted activity',ylabel='Uncertainty')
ax_joint.spines['top'].set_visible(False)
ax_joint.spines['right'].set_visible(False)
g.legend(frameon=False)
for lh in g.legend_.legend_handles: 
    lh.set_alpha(1)
    lh._sizes = [40]

ax_marg_x = fig.add_subplot(gs[0, 0:5], sharex=ax_joint)
g = sns.kdeplot(data=test,x='old_predictions',ax=ax_marg_x,color='lightgrey',multiple='stack',edgecolor=None,alpha=0.8)
sns.kdeplot(data=test,x='new_predictions',ax=ax_marg_x,color='cadetblue',multiple='stack',edgecolor=None,alpha=0.8)
g.set(xlabel=None,ylabel=None,yticklabels=[],yticks=[])
ax_marg_x.xaxis.set_tick_params(labelbottom=False)

ax_marg_y = fig.add_subplot(gs[1:6, 5],sharey=ax_joint)
g = sns.kdeplot(data=test,y='old_uncertainty',ax=ax_marg_y,color='lightgrey',multiple='stack',edgecolor=None,alpha=0.8)
sns.kdeplot(data=test,y='new_uncertainty',ax=ax_marg_y,color='cadetblue',multiple='stack',edgecolor=None,alpha=0.8)
g.set(ylabel=None,xlabel=None,xticklabels=[],xticks=[])
ax_marg_y.yaxis.set_tick_params(labelbottom=False)

plt.subplots_adjust(hspace=0, wspace=0)

if SAVE_FIGURES:
    plt.savefig(f'{FIGURE_DIR}/{NOTEBOOK_ID}-updated_mycocosm_inference.png',transparent=True,bbox_inches='tight',dpi=1200)
```
